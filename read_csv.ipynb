```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_path = './cities.csv'
df = pd.read_csv(csv_path)
```

```python
df = df.set_index('City_ID')
```

```python
df.head()
```

```python
#df.to_html(open('my_data.html', 'w'))
```

```python
#wind speed vs city lat
wind_speed = df['Wind Speed']
city_lat = df['Lat']
```

```python
plt.scatter(city_lat, wind_speed, edgecolor = 'black', color = 'purple')
plt.grid()
plt.title('City Latitude vs. Wind Speed')
plt.ylabel('Wind Speed')
plt.xlabel('City Latitude')
plt.savefig('city_vs_wind.png')
plt.show()
```

```python
#city latitude vs max temp
max_temp = df['Max Temp']
```

```python
plt.scatter(city_lat, max_temp, edgecolor = 'black', color = 'purple')
plt.title('City Latitude vs. Max Temperature')
plt.ylabel('Max Temperature (F)')
plt.xlabel('City Latitude')
plt.grid()
plt.savefig('city_vs_temp.png')
plt.show()
```

```python
#city lat vs humidity
humid = df['Humidity']
```

```python
plt.scatter(city_lat, humid, edgecolor= 'black', color= 'purple')
plt.grid()
plt.title('City Latitude vs. Humidity')
plt.ylabel('Humidity')
plt.xlabel('City Latitude')
plt.savefig('city_vs_humidity.png')
plt.show()
```

```python
#city lat vs cloudiness
cloud = df['Cloudiness']
```

```python
plt.scatter(city_lat, cloud, edgecolor = 'black', color='purple')
plt.grid()
plt.ylabel('Cloudiness')
plt.xlabel('City Latitude')
plt.title('City Latitude vs. Cloudiness')
plt.savefig('city_vs_cloud.png')
plt.show()
```
